```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read and Display Image
img = cv.imread("./images/harry_potter.jfif")

# Convert BGR to RGB Color Space
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

print(img.shape)
print(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
# Accesing Files from their path and Displaying them

import os 

images = os.listdir("./images")

imgjfif = []
for f in images:
    if(f.endswith(".jfif")):
        print(f)
        img = cv.imread("./images/"+f)
        img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
        imgjfif.append(img)
        plt.imshow(img)
        plt.title(img.shape)
        plt.show()
```

```python
pic_12 = np.hstack((imgjfif[0], imgjfif[1]))
pic_34 = np.hstack((imgjfif[2], imgjfif[3]))
collage = np.vstack((pic_12, pic_34))
plt.imshow(collage)
plt.show()
```

```python
# Resizing an image in different proportions

re_image = cv.imread("./images/harry_potter.jfif")

half = cv.resize(re_image, (0,0), fx=0.5, fy=0.5)
bigger = cv.resize(re_image, (1050,1610))
interpolate = cv.resize(re_image, (780,540), fx=0.5, fy=0.5, interpolation=cv.INTER_NEAREST)

count = 4
img_titles = ["Original", "Half", "Bigger", "Interpolate"]
images = [re_image, half, bigger, interpolate]

for i in range(count):
    plt.subplot(2, 2, i+1)
    plt.title(img_titles[i])
    plt.imshow(images[i])
plt.show()
```

```python
# Eroding an Image
img_Ero = cv.imread("./images/sirius_black.jpg")

kernel = np.ones((5, 5), np.uint8)
img_Ero = cv.erode(img_Ero, kernel)

cv.imshow('Eroded Image', img_Ero)
```

```python
# Bitwise Operations
# Bitwise AND      
im1 = cv.imread("./images/bitwise_img1.png")
im2 = cv.imread("./images/bitwise_img2.png")

res = cv.bitwise_and(im1, im2, mask = None)

cv.imshow('Bitiwse AND', res)

if(cv.waitKey(0) & 0xff == 27):
    cv.destroyAllWindows()
```

```python
# Bitwise OR
res_or = cv.bitwise_or(im1, im2, mask = None)

cv.imshow('Bitiwse OR', res_or)

if(cv.waitKey(0) & 0xff == 27):
    cv.destroyAllWindows()
```

```python
# Bitwise XOR
res_xor = cv.bitwise_xor(im1, im2, mask = None)

cv.imshow('Bitiwse XOR', res_xor)

if(cv.waitKey(0) & 0xff == 27):
    cv.destroyAllWindows()
```

```python
# Bitwise NOT
res_notImg1 = cv.bitwise_not(im1, mask = None)
res_notImg2 = cv.bitwise_not(im2, mask = None)

cv.imshow('Bitiwse NOT Image 1', res_notImg1)
cv.imshow('Bitiwse NOT Image 2', res_notImg2)

if(cv.waitKey(0) & 0xff == 27):
    cv.destroyAllWindows()
```

```python
# Printing Plain colors
blue = np.zeros((300, 300, 3), dtype="uint8")
blue[:, :, 2] = 255

green = np.zeros((300, 300, 3), dtype="uint8")
green [:, :, 1] = 255

plt.imshow(blue)
plt.show()

plt.imshow(green)
plt.show()
```

```python
# Image Collage from colors
blue_green = np.hstack([blue, green])
green_blue = blue_green[:,::-1,:]

img = np.vstack((blue_green, green_blue))
plt.imshow(img)
plt.show()
```
